# Threshold Sensitivity Analysis

```python
import os

# Get current directory
dir_path = os.getcwd()  # in an IPython notebook you can also use dir_path = pwd()

# Folders to ignore
ignore_dirs = {'Figures', 'Anonymized','LMEs','.ipynb_checkpoints'}

# Pull company names from folder names
company_names = [
    name for name in os.listdir(dir_path)
    if os.path.isdir(os.path.join(dir_path, name))
    and name not in ignore_dirs
]

print(company_names)
print(dir_path)
```

```python
import os
import datetime
import pandas as pd

# Define your analysis window
start_dt = pd.to_datetime('2023-07-01')
end_dt   = pd.to_datetime('2023-09-30 23:59:59')

# Define a range of threshold values for sensitivity analysis
threshold_values = [i/100 for i in range(0, 101, 5)]  # Range from 0.0 to 1.0 with a step of 0.05

for threshold in threshold_values:
    print(f"Running analysis for threshold: {threshold}")
    for company_name in company_names:
        # Create company directory path
        company_dir_path = os.path.join(dir_path, company_name)
        
        # Use os.path.join to construct the full paths for the CSV files
        energy_file_path = os.path.join(company_dir_path, 'energy_data.csv')
        lme_file_path = os.path.join(company_dir_path, 'lme_data.csv')
        
        # Load the energy and LME data CSV files
        energy_df = pd.read_csv(energy_file_path)
        lme_df = pd.read_csv(lme_file_path)
        
        # Step 1: Convert 'DateTime' column to datetime format for easier handling of dates
        energy_df['DateTime'] = pd.to_datetime(energy_df['DateTime'])
        lme_df['DateTime'] = pd.to_datetime(lme_df['DateTime'])

         # ← new: restrict both datasets to Jul 1–Sep 30, 2023
        energy_df = energy_df[(energy_df['DateTime'] >= start_dt) & (energy_df['DateTime'] <= end_dt)]
        lme_df    = lme_df   [(lme_df   ['DateTime'] >= start_dt) & (lme_df   ['DateTime'] <= end_dt)]
        
        
        # Step 2: Merge the two dataframes on the 'DateTime' column
        df = pd.merge(energy_df, lme_df, on='DateTime', how='left')
        
        # Step 3: Extract additional columns and calculate derived metrics
        df['Hour'] = df['DateTime'].dt.hour
        df['WeekDay'] = df['DateTime'].dt.weekday
        df['Month'] = df['DateTime'].dt.month
        df['Day'] = df['DateTime'].dt.day
        df['energy'] = df['MWh']
        df['daily_max_energy'] = df.groupby(df['DateTime'].dt.date)['energy'].transform('max')
        
        # New Step: Identify and replace anomalously high energy readings based on the previous day's maximum energy
        def replace_outliers(df):
            grouped = df.groupby(df['DateTime'].dt.date)
            for date, group in grouped:
                prev_day = date - datetime.timedelta(days=1)
                
                # Get the previous day's max energy use
                if prev_day in grouped.groups:
                    prev_max_energy = grouped.get_group(prev_day)['daily_max_energy'].iloc[0]
                else:
                    prev_max_energy = 0  # Default to zero if no previous day's data
                
                # Define outlier threshold based on the previous day's max energy (e.g., 1.5x)
                outlier_threshold = 1.5 * prev_max_energy
                outliers = group['energy'] > outlier_threshold
                
                # Replace outliers with the nearest non-outlier value within the group
                if outliers.any():
                    for idx in group.index[outliers]:
                        # Find the nearest non-outlier value within the group
                        nearest_idx = group.loc[group.index < idx, 'energy'][~outliers].last_valid_index()
                        if pd.isna(nearest_idx):
                            nearest_idx = group.loc[group.index > idx, 'energy'][~outliers].first_valid_index()
                        # Replace the outlier value with the nearest valid value
                        df.at[idx, 'energy'] = df.at[nearest_idx, 'energy'] if nearest_idx is not None else group['daily_max_energy'].iloc[0]
            return df
        
        # Apply the outlier replacement function
        df = replace_outliers(df)
    
        # Step 4: Calculate DR_underclock_threshold and DR_flag using the loop variable threshold
        df['daily_max_energy'] = df.groupby(df['DateTime'].dt.date)['energy'].transform('max')
        df['DR_underclock_threshold'] = threshold * df['daily_max_energy']
        df['DR_flag'] = df['energy'] < df['DR_underclock_threshold']  # Re-added this line to calculate DR_flag
    
        # Step 5: Calculate DR_percent for each hour, using the daily average value
        df['energy_daily_avg_no_dr'] = df.groupby(df['DateTime'].dt.date)['energy'].transform(
            lambda x: x[~df.loc[x.index, 'DR_flag']].mean()
        )
        df['energy_savings'] = df['energy_daily_avg_no_dr'] - df['energy']
        df['DR_percent'] = df['energy_savings'] / df['energy_daily_avg_no_dr'] * 100
        df['DR_percent'] = df['DR_percent'].clip(lower=0)
        
        # Step 6: Updated function to detect curtailment periods using DR_flag
        def detect_curtailment(df, threshold_start_end=3):
            df['curtailment_start'] = False
            df['curtailment_end'] = False
            df['DR_active'] = False
            
            for index in range(1, len(df) - 1):
                if df.iloc[index]['DR_flag']:
                    # If DR_flag is True, mark the start of curtailment if the previous row was not in curtailment
                    if not df.iloc[index-1]['DR_flag']:
                        for j in range(index-1, -1, -1):
                            if df.iloc[j]['DR_percent'] < threshold_start_end:
                                df.iloc[j, df.columns.get_loc('curtailment_start')] = True
                                df.iloc[j, df.columns.get_loc('DR_active')] = True
                                break
                    df.iloc[index, df.columns.get_loc('DR_active')] = True
                    # If the next row is not curtailing, mark the end of curtailment
                    if not df.iloc[index+1]['DR_flag']:
                        for j in range(index+1, len(df)):
                            if df.iloc[j]['DR_percent'] < threshold_start_end:
                                df.iloc[j, df.columns.get_loc('curtailment_end')] = True
                                df.iloc[j, df.columns.get_loc('DR_active')] = True
                                break
            return df
        
        # Apply the function to the entire dataset
        df = detect_curtailment(df)
        
        # Step 7: Additional pass to ensure all points between curtailment_start and curtailment_end have DR_active = True
        curtailment_active = False
        for index in range(len(df)):
            if df.iloc[index]['curtailment_start']:
                curtailment_active = True
            if curtailment_active:
                df.iloc[index, df.columns.get_loc('DR_active')] = True
            if df.iloc[index]['curtailment_end']:
                curtailment_active = False
        
        # Step 8: Edge Case Handling - Check for days with daily max energy close to zero
        daily_max_energy_threshold = 1  # Adjust this threshold as needed
        grouped = df.groupby(df['DateTime'].dt.date)
        
        for date, group in grouped:
            if group['daily_max_energy'].iloc[0] < daily_max_energy_threshold:
                # Set DR_flag and DR_active to True for all entries of that day
                df.loc[group.index, 'DR_flag'] = True
                df.loc[group.index, 'DR_active'] = True
                
                # Get one energy_daily_avg_no_dr value from the previous and next days
                prev_day = date - datetime.timedelta(days=1)
                next_day = date + datetime.timedelta(days=1)
                
                prev_avg = grouped.get_group(prev_day)['energy_daily_avg_no_dr'].iloc[0] if prev_day in grouped.groups else 0
                next_avg = grouped.get_group(next_day)['energy_daily_avg_no_dr'].iloc[0] if next_day in grouped.groups else 0
                avg_energy_no_dr = (prev_avg + next_avg) / 2
                
                # Update energy_savings and emissions calculations for the curtailment day
                df.loc[group.index, 'energy_daily_avg_no_dr'] = avg_energy_no_dr
                df.loc[group.index, 'energy_savings'] = avg_energy_no_dr - df.loc[group.index, 'energy']
                df.loc[group.index, 'DR_percent'] = df.loc[group.index, 'energy_savings'] / avg_energy_no_dr * 100
        
        # Step 9: Proceed with the rest of your calculations (energy savings, DR percent, emissions, etc.)
        
        # Calculate induced_emissions as energy * LME
        df['induced_emissions'] = df['energy'] * df['LME']
        
        # Calculate avoided_emissions as energy_savings * LME, only if DR_active is True
        df['avoided_emissions'] = df.apply(lambda row: row['energy_savings'] * row['LME'] if row['DR_active'] else 0, axis=1)
        
        # Step 10: Adjust avoided_emissions to zero if negative only if LME is positive
        df['avoided_emissions'] = df.apply(
            lambda row: 0 if row['avoided_emissions'] < 0 and row['LME'] > 0 else row['avoided_emissions'], axis=1
        )
        
        # Calculate net_emissions as induced_emissions - avoided_emissions
        df['net_emissions'] = df['induced_emissions'] - df['avoided_emissions']
        
        # Save results for each threshold
        results_path = os.path.join(company_dir_path, f'curtailment_results_threshold_{int(threshold * 100)}.csv')
        df.to_csv(results_path, index=False)
        print(f"Saved results for threshold {threshold}: {results_path}")
```

```python
import os
import glob
import pandas as pd

# Initialize a list to collect summary records
summary_records = []

# Loop over each company directory
for company_name in company_names:
    company_dir_path = os.path.join(dir_path, company_name)
    
    # Using glob to match files following the naming pattern
    result_files = glob.glob(os.path.join(company_dir_path, 'curtailment_results_threshold_*.csv'))
    
    for result_file in result_files:
        # Extract the threshold value from the file name.
        # File names are expected to be like "curtailment_results_threshold_70.csv" for a threshold of 0.70.
        file_name = os.path.basename(result_file)
        threshold_str = file_name.replace('curtailment_results_threshold_', '').replace('.csv', '')
        threshold_val = float(threshold_str) / 100.0
        
        # Read the results CSV file
        df = pd.read_csv(result_file)
        
        # Compute summary metrics. You can modify or extend these as needed.
        mean_DR_percent = df['DR_percent'].mean()
        total_energy_savings = df['energy_savings'].sum()
        total_net_emissions = df['net_emissions'].sum()
        
        # Append a summary record for the company at this threshold value
        summary_records.append({
            'company': company_name,
            'threshold': threshold_val,
            'mean_DR_percent': mean_DR_percent,
            'total_energy_savings': total_energy_savings,
            'total_net_emissions': total_net_emissions
        })

# Create a DataFrame from the summary records
summary_df = pd.DataFrame(summary_records)

# Display the summary DataFrame
print(summary_df)

# Save the summary DataFrame to a CSV file for further analysis or visualization
summary_path = os.path.join(dir_path, 'sensitivity_summary.csv')
summary_df.to_csv(summary_path, index=False)
print(f"Sensitivity summary saved to: {summary_path}")
```

```python
import os

# Get current directory
dir_path = os.getcwd()  # in an IPython notebook you can also use dir_path = pwd()

# Folders to ignore
ignore_dirs = {'Figures', 'Anonymized','LMEs','.ipynb_checkpoints'}

# Pull company names from folder names
company_names = [
    name for name in os.listdir(dir_path)
    if os.path.isdir(os.path.join(dir_path, name))
    and name not in ignore_dirs
]

print(company_names)
print(dir_path)
```

```python
import os
import glob
import pandas as pd

# List of company names and base directory path should be defined in your environment.
# For example:
# company_names = [CompanyB', 'CompanyC']
# dir_path = '/path/to/your/companies'

summary_records = []

# Loop over companies
for company in company_names:
    company_dir_path = os.path.join(dir_path, company)
    # Use glob to find all CSV files that match the naming pattern for sensitivity results.
    # Files are assumed to have names like "curtailment_results_threshold_XX.csv"
    result_files = glob.glob(os.path.join(company_dir_path, 'curtailment_results_threshold_*.csv'))
    
    for result_file in result_files:
        # Extract the threshold value from the filename.
        # For example, a file named "curtailment_results_threshold_70.csv" indicates threshold 0.70.
        file_name = os.path.basename(result_file)
        threshold_str = file_name.replace('curtailment_results_threshold_', '').replace('.csv', '')
        try:
            threshold_val = float(threshold_str) / 100.0
        except ValueError:
            # Skip files that don't follow the naming convention
            continue

        # Read the CSV results
        df = pd.read_csv(result_file)

        # Compute summary metrics (customize this list as needed)
        mean_DR_percent = df['DR_percent'].mean() if 'DR_percent' in df.columns else None
        total_energy_savings = df['energy_savings'].sum() if 'energy_savings' in df.columns else None
        total_net_emissions = df['net_emissions'].sum() if 'net_emissions' in df.columns else None

        # Append the record to our list
        summary_records.append({
            'company': company,
            'threshold': threshold_val,
            'mean_DR_percent': mean_DR_percent,
            'total_energy_savings': total_energy_savings,
            'total_net_emissions': total_net_emissions
        })

# Create the aggregated DataFrame
summary_all = pd.DataFrame(summary_records)

# Optional: restrict to the threshold range [0.2, 1.0] if desired.
summary_all = summary_all[(summary_all['threshold'] >= 0.2) & (summary_all['threshold'] <= 1.0)]

# Display the aggregated summary DataFrame
print("Aggregated Summary Across All Companies:")
print(summary_all.head())

# You can now proceed with groupby analysis to compute the inflection (elbow) point for each company.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def kneedle_max_distance(x, y, visualize=False):
    """
    Knee detection via max perpendicular distance, but without normalizing x,
    only normalizing y to [0,1]. Returns the knee in the original x-scale.
    """
    x = np.asarray(x)
    y = np.asarray(y)
    
    # Normalize only y to [0, 1]
    y_norm = (y - np.min(y)) / (np.max(y) - np.min(y))
    
    # Line endpoints in (x, y_norm) space
    x0, yn0 = x[0], y_norm[0]
    x1, yn1 = x[-1], y_norm[-1]
    
    # Line coefficients a*x + b*y + c = 0
    a = yn1 - yn0
    b = -(x1 - x0)
    c = x1*yn0 - yn1*x0
    
    # Compute perpendicular distances
    distances = np.abs(a * x + b * y_norm + c) / np.sqrt(a**2 + b**2)
    
    idx = np.argmax(distances)
    knee_x = x[idx]
    
    if visualize:
        plt.figure(figsize=(8,5))
        plt.plot(x, y_norm, 'o-', label="Normalized DR (%)")
        plt.plot([x0, x1], [yn0, yn1], 'r--', label="Baseline")
        plt.scatter(x[idx], y_norm[idx], color='green', s=100, label="Knee")
        plt.xlabel("Threshold")
        plt.ylabel("Normalized Mean DR (%)")
        #plt.title("Kneedle (Perp. Distance) without x‑Normalization")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        
        # ← save the current figure to file before showing
        plt.savefig('kneedle_plot.png', dpi=300, bbox_inches='tight')
        plt.show()
    
    return knee_x

# Example usage:
# Assume summary_all is your aggregated DataFrame with columns:
# 'company', 'threshold', 'mean_DR_percent', etc.
# Restrict to thresholds between 0.2 and 1.0 (or another appropriate range).
filtered_summary = summary_all[(summary_all['threshold'] >= 0) & (summary_all['threshold'] <= 1.0)]

knee_data = []
for company, group in filtered_summary.groupby('company'):
    group = group.sort_values('threshold')
    x = group['threshold'].to_numpy()
    # Choose your metric of interest. For example, we use 'mean_DR_percent'
    y = group['mean_DR_percent'].to_numpy()
    knee = kneedle_max_distance(x, y, visualize=True)
    knee_data.append({'company': company, 'knee_threshold': knee})

knee_df = pd.DataFrame(knee_data)
print("Knee Thresholds for Each Company:")
print(knee_df)

# Optionally, compute summary statistics across companies.
stats = knee_df['knee_threshold'].describe()
print("\nSummary Statistics for the Knee Thresholds:")
print(stats)
```
