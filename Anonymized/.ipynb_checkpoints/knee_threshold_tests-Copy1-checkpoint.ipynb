```python
import os
import glob
import pandas as pd
import numpy as np

# List of facilities with their knee thresholds
knee_list = [
    {'company': 'CPower',         'knee_threshold': 0.9},
    {'company': 'Calvert_City_1', 'knee_threshold': 0.95},
    {'company': 'Calvert_City_2', 'knee_threshold': 0.9},
    {'company': 'Calvert_City_3', 'knee_threshold': 0.8},
    {'company': 'CleanSpark',     'knee_threshold': 0.95},
    {'company': 'Cormint',        'knee_threshold': 0.75},
    {'company': 'Cottonwood_1',   'knee_threshold': 0.85},
    {'company': 'Dalton_1',       'knee_threshold': 0.9},
    {'company': 'Dalton_2',       'knee_threshold': 0.9},
    {'company': 'Dalton_3',       'knee_threshold': 0.9},
    {'company': 'Denton_DTO1',    'knee_threshold': 0.9},
    {'company': 'Foreman',        'knee_threshold': 0.85},
    {'company': 'Grand_Forks_1',  'knee_threshold': 0.85},
    {'company': 'IrisEnergy',     'knee_threshold': 0.85},
    {'company': 'Lancium',        'knee_threshold': 0.85},
    {'company': 'Marble_1',       'knee_threshold': 0.95},
    {'company': 'Marble_2',       'knee_threshold': 0.95},
    {'company': 'Mining Works',   'knee_threshold': 0.85},
    {'company': 'SATOtech',       'knee_threshold': 0.9},
    {'company': 'Soluna',         'knee_threshold': 0.85},
    {'company': 'StandardPower',  'knee_threshold': 0.9},
]

results = []

for idx, info in enumerate(knee_list, start=1):
    comp = info['company']
    knee_thr = info['knee_threshold']
    pattern = f"miner_{idx}_curtailment_results_*.csv"
    files = glob.glob(pattern)
    print(files)
    
    dr_records = []
    for fn in files:
        thr_val = int(fn.split('_')[-1].split('.')[0]) / 100
        df = pd.read_csv(fn)
        dr_hours = df['DR_active'].sum()
        dr_records.append((thr_val, dr_hours))
        daily_max_energy = df['daily_max_energy'].max()
        avoided_emissions = df['avoided_emissions'].sum()
        induced_emissions = df['induced_emissions'].sum()
    
    if not dr_records:
        continue
    
    df_dr = pd.DataFrame(dr_records, columns=['threshold', 'dr_hours'])
    max_hours = len(df) - dr_hours
    # DR at knee threshold
    knee_row = df_dr[df_dr['threshold'] == knee_thr]
    knee_hours = knee_row['dr_hours'].iloc[0] if not knee_row.empty else np.nan
    # relative percentage
    relative_pct = (knee_hours / max_hours * 100) if max_hours > 0 else np.nan
    
    results.append({
        'company': comp,
        'knee_threshold': knee_thr,
        'DR_hours_at_knee': knee_hours,
        'max_DR_hours': max_hours,
        'relative_DR_pct': relative_pct,
        'daily_max_energy': daily_max_energy,
        'avoided_emissions': avoided_emissions,
        'induced_emissions:': induced_emissions
    })

df_results = pd.DataFrame(results)
df_results.sort_values('knee_threshold', ascending=False, inplace=True)
df_results
```

```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, levene, mannwhitneyu

# assume df_results is already in your namespace
# e.g. df_results = pd.read_csv('your_results.csv')

# 1) Split into the two groups
low_group  = df_results[df_results['knee_threshold'] <= 0.85]
high_group = df_results[df_results['knee_threshold'] >= 0.90]

metric = 'relative_DR_pct'

low_vals  = low_group[metric].dropna()
high_vals = high_group[metric].dropna()

# 2) Check assumptions

# 2a) Normality (ShapiroWilk)
print("Shapiro normality test:")
print("  low  :", shapiro(low_vals))
print("  high :", shapiro(high_vals))

# 2b) Variance homogeneity (Levenes test)
print("\nLevenes test for equal variances:")
print("  stat, p =", levene(low_vals, high_vals))

# 3) Run Welchs t-test (doesnt assume equal variances)
t_stat, p_val = ttest_ind(low_vals, high_vals, equal_var=False)
print(f"\nWelchs t-test on {metric!r}: t = {t_stat:.3f}, p = {p_val:.3f}")

# 4) If normality/variance assumptions fail, fall back to MannWhitney U
u_stat, p_mw = mannwhitneyu(low_vals, high_vals, alternative='two-sided')
print(f"MannWhitney U on {metric!r}: U = {u_stat:.3f}, p = {p_mw:.3f}")
```

```python
for metric in ['relative_DR_pct', 'avoided_emissions', 'daily_max_energy', 'max_DR_hours', 'DR_hours_at_knee']:
    low_vals  = low_group[metric].dropna()
    high_vals = high_group[metric].dropna()
    t, p = ttest_ind(low_vals, high_vals, equal_var=False)
    print(f"{metric}: t={t:.2f}, p={p:.3f}")
```

```python
import numpy as np

def cohens_d(a, b):
    sd_pooled = np.sqrt((a.var(ddof=1) + b.var(ddof=1)) / 2)
    return (a.mean() - b.mean()) / sd_pooled

for metric in ['relative_DR_pct', 'DR_hours_at_knee']:
    d = cohens_d(
        high_group[metric].dropna(), 
        low_group[metric].dropna()
    )
    print(f"{metric}: Cohens d = {d:.2f}")
```

The amount of active DR hours is statistically significant across groups, but not the total avoided emissions or induced emissions. This means that simply increasing your demand response is alone insufficient to maximize avoided or reduce your emissions. With the regression model, we found that avoided emissions could instead be partly explaiend by equal parts LME variability and curtailment regularity. These two separate statistical findings show that our facilities could benefit from strategic placement and LME hourly alignment to maximize benefit to the grid and carbon reduction goals.

```python
import os
import glob
import pandas as pd
import numpy as np

def calculate_curtailment_magnitude(df):
    """
    Calculates the % curtailment magnitude as:
      100 * (1 - (avg_energy_in_lowest_quartile_of_DR_hours / baseline_avg_no_DR))
    """
    # 1) grab just the energy during DRactive hours
    curtailment_hours = df.loc[df['DR_active'], 'energy'].dropna()

    if curtailment_hours.empty:
        # no DR hours => zero curtailment magnitude
        return 0.0

    # 2) find the 25th percentile of those DRhour energies
    q25 = np.percentile(curtailment_hours, 25)
    low_q = curtailment_hours[curtailment_hours <= q25]

    # 3) average of that bottom quartile (or fall back to min if somehow empty)
    avg_low_q = low_q.mean() if not low_q.empty else curtailment_hours.min()

    # 4) baseline: mean energy during non-DR hours
    baseline = df.loc[~df['DR_active'], 'energy'].mean()
    if pd.isna(baseline) or baseline == 0:
        # can't divide by zero or missing
        return 0.0

    # 5) percent curtailment magnitude
    return 100.0 * (1.0 - (avg_low_q / baseline))

# 1) List of facilities + knee thresholds
knee_list = [
    {'company': 'CPower',         'knee_threshold': 0.9},
    {'company': 'Calvert_City_1', 'knee_threshold': 0.95},
    {'company': 'Calvert_City_2', 'knee_threshold': 0.9},
    {'company': 'Calvert_City_3', 'knee_threshold': 0.8},
    {'company': 'CleanSpark',     'knee_threshold': 0.95},
    {'company': 'Cormint',        'knee_threshold': 0.75},
    {'company': 'Cottonwood_1',   'knee_threshold': 0.85},
    {'company': 'Dalton_1',       'knee_threshold': 0.9},
    {'company': 'Dalton_2',       'knee_threshold': 0.9},
    {'company': 'Dalton_3',       'knee_threshold': 0.9},
    {'company': 'Denton_DTO1',    'knee_threshold': 0.9},
    {'company': 'Foreman',        'knee_threshold': 0.85},
    {'company': 'Grand_Forks_1',  'knee_threshold': 0.85},
    {'company': 'IrisEnergy',     'knee_threshold': 0.85},
    {'company': 'Lancium',        'knee_threshold': 0.85},
    {'company': 'Marble_1',       'knee_threshold': 0.95},
    {'company': 'Marble_2',       'knee_threshold': 0.95},
    {'company': 'Mining Works',   'knee_threshold': 0.85},
    {'company': 'SATOtech',       'knee_threshold': 0.9},
    {'company': 'Soluna',         'knee_threshold': 0.85},
    {'company': 'StandardPower',  'knee_threshold': 0.9},
]

all_records = []

for idx, info in enumerate(knee_list, start=1):
    comp = info['company']
    knee_thr = info['knee_threshold']
    pattern = f"miner_{idx}_curtailment_results_*.csv"
    files = glob.glob(pattern)
    if not files:
        continue

    for fn in files:
        # extract threshold from filename
        thr_val = int(os.path.splitext(fn)[0].split('_')[-1]) / 100

        # load data
        df = pd.read_csv(fn, parse_dates=['DateTime'])
        
        # basic counts
        total_hours     = len(df)
        dr_active_hours = df['DR_active'].sum()
        uptime_hours    = total_hours - dr_active_hours
        
        # 1. Uptime %  (non-DR hours)
        uptime_pct = uptime_hours / total_hours * 100 if total_hours>0 else np.nan
        
        # 2/3. Emissions sums
        avoided_em = df['avoided_emissions'].sum()
        induced_em = df['induced_emissions'].sum()

        # 4. Daily max energy
        max_energy = df['daily_max_energy'].max()

        # 5. DR relative %  
        dr_relative_pct = dr_active_hours / total_hours * 100 if total_hours>0 else np.nan

        # 6. LME variability
        lme_var = df['LME'].std() if 'LME' in df else np.nan

        # 7. Energy savings & emission efficiency
        energy_savings = df['energy_savings'].sum() if 'energy_savings' in df else np.nan
        emission_eff   = (avoided_em / energy_savings
                          if energy_savings and energy_savings>0
                          else np.nan)

        # 8. Curtailment regularity
        if 'curtailment_start' in df and 'DateTime' in df:
            starts = df.loc[df['curtailment_start'], 'DateTime'].dt.time
            if len(starts)>1:
                secs = [t.hour*3600 + t.minute*60 + t.second for t in starts]
                curtail_reg = np.std(secs) / 3600.0
            else:
                curtail_reg = np.nan
        else:
            curtail_reg = np.nan

        # 9. Curtailment magnitude
        curtail_mag = calculate_curtailment_magnitude(df)

        # 10. Facility ID (assumed constant in file)
        facility_id = df['Facility ID'].iloc[0] if 'Facility ID' in df else comp

        # 11. (later) knee threshold DR hours + relative %
        #   well pick them out from the perthreshold list below

        all_records.append({
            'Company': comp,
            'Facility ID': facility_id,
            'Threshold': thr_val,
            'Uptime (%)': uptime_pct,
            'DR Hours': dr_active_hours,
            'DR Relative (%)': dr_relative_pct,
            'Avoided Emissions (tCO)': avoided_em,
            'Induced Emissions (tCO)': induced_em,
            'Daily Max Energy (MWh)': max_energy,
            'LME Variability': lme_var,
            'Energy Savings (tCO)': energy_savings,
            'Emission Efficiency': emission_eff,
            'Curtailment Regularity (hr)': curtail_reg,
            'Curtailment Magnitude (%)': curtail_mag,
            'Knee Threshold': knee_thr,
        })

# consolidate to a DataFrame
df_all = pd.DataFrame(all_records)

# 13. New derived metrics
df_all['Emissions Ratio'] = (
    df_all['Avoided Emissions (tCO)']
    / df_all['Induced Emissions (tCO)']
).replace([np.inf, -np.inf], np.nan)

df_all['Normalized Avoided Emissions (tCO/MWh)'] = (
    df_all['Avoided Emissions (tCO)']
    / df_all['Daily Max Energy (MWh)']
).replace([np.inf, -np.inf], np.nan)


df_all

# # 12. Extract DR hours at knee threshold, per facility
# #     we assume each (Company, Knee Threshold) pair is unique:
# knee_summary = (
#     df_all
#     .query("Threshold == Knee Threshold")
#     .loc[:, ['Company', 'DR Hours', 'DR Relative (%)']]
#     .rename(columns={
#         'DR Hours': 'DR at Knee (hrs)',
#         'DR Relative (%)': 'DR at Knee (%)'
#     })
# )

# # merge back onto full table if you like, e.g.:
# df_final = df_all.merge(
#     knee_summary,
#     on='Company',
#     how='left'
# )

# # 13. New derived metrics
# df_final['Emissions Ratio'] = (
#     df_final['Avoided Emissions (tCO)']
#     / df_final['Induced Emissions (tCO)']
# ).replace([np.inf, -np.inf], np.nan)

# df_final['Normalized Avoided Emissions (tCO/MWh)'] = (
#     df_final['Avoided Emissions (tCO)']
#     / df_final['Daily Max Energy (MWh)']
# ).replace([np.inf, -np.inf], np.nan)

# # sort & save
# df_final.sort_values(['Company','Threshold'], inplace=True)
# df_final.to_csv('facility_curtailment_metrics.csv', index=False)

# # show
# print(df_final.head(10))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  assume uptime_df is already in memory with columns:
#    'Facility ID', 'Uptime (%)',
#    'Avoided Emissions (tons CO2)',
#    'Induced Emissions (tons CO2)',
#    'Threshold'


# 1. Compute the 25th and 50th percentiles of Threshold
uptime_df = df_all
thresholds = uptime_df['Knee Threshold']
q1 = np.percentile(thresholds, 25)
q2 = np.percentile(thresholds, 50)

low_label  = f'{int(q1*100)}%'
high_label = f'{int(q2*100)}%'

# 2. Split into two groups
groups = {
    low_label:  uptime_df[uptime_df['Knee Threshold'] <= q1],
    high_label: uptime_df[uptime_df['Knee Threshold'] >= q2]
}

# 3. Define your four metrics
metrics = [
    ('Curtailment Magnitude (%)',     'Curtailment Magnitude (%)'),
    ('Uptime (%)',                    'Uptime (%)'),
    ('Induced Emissions (tCO)', 'Induced Emissions'),
    ('Emissions Ratio', 'Emissions Ratio')
]

# 4. Make a 22 grid (one subplot per metric) and flatten it
fig, axes = plt.subplots(1, 4, figsize=(12, 8), constrained_layout=True)
axes = axes.flatten()

# 5. Loop over axes and metrics
for ax, (col, title) in zip(axes, metrics):
    data = [
        groups[low_label][col].values,
        groups[high_label][col].values
    ]
    ax.boxplot(data, tick_labels=[low_label, high_label])
    ax.set_title(title)
    ax.set_ylabel(title)
#plt.savefig('t-test_threshold_results.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, levene, mannwhitneyu

# assume df_results is already in your namespace
# e.g. df_results = pd.read_csv('your_results.csv')

stats = uptime_df['Threshold'].describe()
q1     = stats['25%']
median = stats['50%']
q3     = stats['75%']
mean   = stats['mean']

# 1) Split into the two groups
low_group  = uptime_df[uptime_df['Threshold'] < 0.9]
high_group = uptime_df[uptime_df['Threshold'] >= 0.9]

metric = 'DR Relative (%)'

low_vals  = low_group[metric].dropna()
high_vals = high_group[metric].dropna()

# 2) Check assumptions

# 2a) Normality (ShapiroWilk)
print("Shapiro normality test:")
print("  low  :", shapiro(low_vals))
print("  high :", shapiro(high_vals))

# 2b) Variance homogeneity (Levenes test)
print("\nLevenes test for equal variances:")
print("  stat, p =", levene(low_vals, high_vals))

# 3) Run Welchs t-test (doesnt assume equal variances)
t_stat, p_val = ttest_ind(low_vals, high_vals, equal_var=False)
print(f"\nWelchs t-test on {metric!r}: t = {t_stat:.3f}, p = {p_val:.3f}")

# 4) If normality/variance assumptions fail, fall back to MannWhitney U
u_stat, p_mw = mannwhitneyu(low_vals, high_vals, alternative='two-sided')
print(f"MannWhitney U on {metric!r}: U = {u_stat:.3f}, p = {p_mw:.3f}")
```

```python
# Add these imports to the top of your script
import scipy.stats as stats
from scipy.stats import ttest_ind, mannwhitneyu

# --- Add these Helper Functions (if needed) ---
def cohens_d(a, b):
    """Calculates Cohen's d for independent samples."""
    sd_pooled = np.sqrt((a.var(ddof=1) + b.var(ddof=1)) / 2)
    return (a.mean() - b.mean()) / sd_pooled

def rank_biserial_correlation(U, n1, n2):
    """Calculates Rank-Biserial Correlation for Mann-Whitney U."""
    return 1 - (2 * U) / (n1 * n2)

# --- Your existing metrics list and dataframes ('low_group', 'high_group') are assumed to be defined ---
metrics = [
    'Uptime (%)',
    'Avoided Emissions (tCO)',
    'Daily Max Energy (MWh)',
    'Induced Emissions (tCO)',
    'LME Variability',
    'Curtailment Magnitude (%)',
    'Emissions Ratio',
    'Curtailment Regularity (hr)'
]


# --- Main Analysis Loop ---
for metric in metrics:
    low_vals  = low_group[metric].dropna()
    high_vals = high_group[metric].dropna()

    print(f"\n{'='*60}")
    print(f"Analysis for Metric: {metric}")
    print(f"{'='*60}")

    # 1. VISUAL NORMALITY CHECK (Histogram and Q-Q Plot)
    fig, ax = plt.subplots(2, 2, figsize=(10, 8))
    fig.suptitle(f'Normality Check for "{metric}"', fontsize=16)
    ax[0, 0].hist(low_vals, bins='auto', edgecolor='black')
    ax[0, 0].set_title('Histogram: low group (knee  0.85)')
    ax[0, 1].hist(high_vals, bins='auto', color='salmon', edgecolor='black')
    ax[0, 1].set_title('Histogram: high group (knee  0.90)')
    stats.probplot(low_vals, dist="norm", plot=ax[1, 0])
    ax[1, 0].set_title('Q-Q Plot: low group')
    stats.probplot(high_vals, dist="norm", plot=ax[1, 1])
    ax[1, 1].set_title('Q-Q Plot: high group')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    # 2. DESCRIPTIVE STATISTICS
    mean_low, std_low   = low_vals.mean(),  low_vals.std(ddof=1)
    mean_high, std_high = high_vals.mean(), high_vals.std(ddof=1)
    print("\n Descriptive Statistics:")
    print(f"  low group  (n={len(low_vals)}): mean = {mean_low:.2f}, SD = {std_low:.2f}")
    print(f"  high group (n={len(high_vals)}): mean = {mean_high:.2f}, SD = {std_high:.2f}")

    # 3. STATISTICAL TESTS
    print("\n Statistical Test Results:")
    # Student's t-test (assumes equal variance)
    t_student, p_student = ttest_ind(low_vals, high_vals, equal_var=True)
    print(f"  - Student's t-test:   t = {t_student:.2f}, p = {p_student:.3f}")
    # Welch's t-test (does NOT assume equal variance)
    t_welch, p_welch = ttest_ind(low_vals, high_vals, equal_var=False)
    print(f"  - Welch's t-test:       t = {t_welch:.2f}, p = {p_welch:.3f}")
    # Mann-Whitney U test (non-parametric)
    U, p_mannwhitney = mannwhitneyu(low_vals, high_vals, alternative='two-sided')
    print(f"  - Mann-Whitney U test:  U = {U:.1f}, p = {p_mannwhitney:.3f}")

    # 4. EFFECT SIZES
    print("\n Effect Sizes:")
    d = cohens_d(high_vals, low_vals)
    print(f"  - Cohens d (for t-tests): {d:.2f}")
    rbc = rank_biserial_correlation(U, len(low_vals), len(high_vals))
    print(f"  - Rank-Biserial Correlation (for U test): {rbc:.2f}")

    # 5. BOXPLOT VISUALIZATION
    plt.figure(figsize=(6, 4))
    plt.boxplot([low_vals, high_vals], labels=['knee  0.85', 'knee  0.90'])
    plt.title(f'Comparison for "{metric}"')
    plt.ylabel(metric)
    plt.tight_layout()
    plt.show()
```
