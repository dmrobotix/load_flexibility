```python
import os
import glob
import pandas as pd

# List of company names and base directory path should be defined in your environment.
# For example:
# company_names = ['Soluna', 'CompanyB', 'CompanyC']
# dir_path = '/path/to/your/companies'

summary_records = []

# Loop over companies
for company in company_names:
    company_dir_path = os.path.join(dir_path, company)
    # Use glob to find all CSV files that match the naming pattern for sensitivity results.
    # Files are assumed to have names like "curtailment_results_threshold_XX.csv"
    result_files = glob.glob(os.path.join(company_dir_path, 'curtailment_results_threshold_*.csv'))
    
    for result_file in result_files:
        # Extract the threshold value from the filename.
        # For example, a file named "curtailment_results_threshold_70.csv" indicates threshold 0.70.
        file_name = os.path.basename(result_file)
        threshold_str = file_name.replace('curtailment_results_threshold_', '').replace('.csv', '')
        try:
            threshold_val = float(threshold_str) / 100.0
        except ValueError:
            # Skip files that don't follow the naming convention
            continue

        # Read the CSV results
        df = pd.read_csv(result_file)

        # Compute summary metrics (customize this list as needed)
        mean_DR_percent = df['DR_percent'].mean() if 'DR_percent' in df.columns else None
        total_energy_savings = df['energy_savings'].sum() if 'energy_savings' in df.columns else None
        total_net_emissions = df['net_emissions'].sum() if 'net_emissions' in df.columns else None

        # Append the record to our list
        summary_records.append({
            'company': company,
            'threshold': threshold_val,
            'mean_DR_percent': mean_DR_percent,
            'total_energy_savings': total_energy_savings,
            'total_net_emissions': total_net_emissions
        })

# Create the aggregated DataFrame
summary_all = pd.DataFrame(summary_records)

# Optional: restrict to the threshold range [0.2, 1.0] if desired.
summary_all = summary_all[(summary_all['threshold'] >= 0.2) & (summary_all['threshold'] <= 1.0)]

# Display the aggregated summary DataFrame
print("Aggregated Summary Across All Companies:")
print(summary_all.head())

# You can now proceed with groupby analysis to compute the inflection (elbow) point for each company.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assume 'summary_all' is a pre-existing DataFrame.
# If you need to test this script, uncomment the following lines
# to create a sample DataFrame.
#
# data = []
# for company in ['A', 'B']:
#     thresholds = np.linspace(0, 1, 50)
#     # Create a concave up curve
#     mean_dr = 100 * (thresholds**2.5) + np.random.rand(50) * 5
#     for t, dr in zip(thresholds, mean_dr):
#         data.append({'company': company, 'threshold': t, 'mean_DR_percent': dr})
# summary_all = pd.DataFrame(data)


def kneedle_max_distance(x, y, visualize=False):
    """
    Knee detection via max perpendicular distance, normalizing both x and y
    to [0,1]. Returns the knee in the original x-scale.
    The linear normalization used preserves the shape (e.g., concavity)
    of the original curve.
    """
    x = np.asarray(x)
    y = np.asarray(y)

    # Normalize both x and y to a [0, 1] range
    x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
    y_norm = (y - np.min(y)) / (np.max(y) - np.min(y))

    # Line endpoints in the normalized (x_norm, y_norm) space
    xn0, yn0 = x_norm[0], y_norm[0]
    xn1, yn1 = x_norm[-1], y_norm[-1]

    # Line coefficients for the line equation a*x + b*y + c = 0
    # This line connects the start and end points in the normalized space.
    a = yn1 - yn0
    b = -(xn1 - xn0)
    c = xn1 * yn0 - yn1 * xn0

    # Compute the perpendicular distance of each point from the line
    distances = np.abs(a * x_norm + b * y_norm + c) / np.sqrt(a**2 + b**2)

    # Find the index of the point with the maximum distance
    idx = np.argmax(distances)
    # Get the knee point in the original x-scale
    knee_x = x[idx]

    if visualize:
        plt.figure(figsize=(8, 5))
        # Plot the normalized data
        plt.plot(x_norm, y_norm, 'o-', label="Normalized DR (%) Curve")
        # Plot the baseline connecting the start and end points
        plt.plot([xn0, xn1], [yn0, yn1], 'r--', label="Baseline")
        # Highlight the detected knee point
        plt.scatter(x_norm[idx], y_norm[idx], color='green', s=100, zorder=5, label="Knee")

        plt.xlabel("Normalized Threshold")
        plt.ylabel("Normalized Mean DR (%)")
        plt.title("Kneedle (Perp. Distance) with X & Y Normalization")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        # Save the current figure to a new file
        plt.savefig('kneedle_plot_normalized.png', dpi=300, bbox_inches='tight')
        plt.show()

    return knee_x

# --- Example usage ---

# Ensure the DataFrame is not empty
if 'summary_all' in locals() and not summary_all.empty:
    # Restrict to thresholds between 0 and 1.0
    filtered_summary = summary_all[(summary_all['threshold'] >= 0) & (summary_all['threshold'] <= 1.0)]

    knee_data = []
    # Group by company and find the knee for each one
    for company, group in filtered_summary.groupby('company'):
        # Sort values by the threshold to ensure the curve is correctly ordered
        group = group.sort_values('threshold')
        x = group['threshold'].to_numpy()
        y = group['mean_DR_percent'].to_numpy()

        # Ensure there are enough points to form a curve
        if len(x) > 2:
            knee = kneedle_max_distance(x, y, visualize=True)
            knee_data.append({'company': company, 'knee_threshold': knee})
        else:
            print(f"Skipping company '{company}' due to insufficient data points.")


    if knee_data:
        knee_df = pd.DataFrame(knee_data)
        print("Knee Thresholds for Each Company:")
        print(knee_df)

        # Optionally, compute summary statistics across all companies.
        stats = knee_df['knee_threshold'].describe()
        print("\nSummary Statistics for the Knee Thresholds:")
        print(stats)
    else:
        print("Could not compute knee thresholds. Check your input data.")
else:
    print("DataFrame 'summary_all' not found or is empty. Please define it.")
```

# THRESHOLD STATISTICS

```python
import os
import glob
import pandas as pd
import numpy as np

def calculate_curtailment_magnitude(df):
    """
    Calculates the % curtailment magnitude as:
      100 * (1 - (avg_energy_in_lowest_quartile_of_DR_hours / baseline_avg_no_DR))
    """
    # 1) grab just the energy during DRactive hours
    curtailment_hours = df.loc[df['DR_active'], 'energy'].dropna()

    if curtailment_hours.empty:
        # no DR hours => zero curtailment magnitude
        return 0.0

    # 2) find the 25th percentile of those DRhour energies
    q25 = np.percentile(curtailment_hours, 25)
    low_q = curtailment_hours[curtailment_hours <= q25]

    # 3) average of that bottom quartile (or fall back to min if somehow empty)
    avg_low_q = low_q.mean() if not low_q.empty else curtailment_hours.min()

    # 4) baseline: mean energy during non-DR hours
    baseline = df.loc[~df['DR_active'], 'energy'].mean()
    if pd.isna(baseline) or baseline == 0:
        # can't divide by zero or missing
        return 0.0

    # 5) percent curtailment magnitude
    return 100.0 * (1.0 - (avg_low_q / baseline))

# 1) List of facilities + knee thresholds
knee_list = [
    {'company': 'CPower',           'knee_threshold': 0.9},
    {'company': 'Calvert_City_1', 'knee_threshold': 0.95},
    {'company': 'Calvert_City_2', 'knee_threshold': 0.9},
    {'company': 'Calvert_City_3', 'knee_threshold': 0.8},
    {'company': 'CleanSpark',       'knee_threshold': 0.95},
    {'company': 'Cormint',         'knee_threshold': 0.75},
    {'company': 'Cottonwood_1',   'knee_threshold': 0.85},
    {'company': 'Dalton_1',       'knee_threshold': 0.9},
    {'company': 'Dalton_2',       'knee_threshold': 0.9},
    {'company': 'Dalton_3',       'knee_threshold': 0.9},
    {'company': 'Denton_DTO1',    'knee_threshold': 0.9},
    {'company': 'Foreman',         'knee_threshold': 0.85},
    {'company': 'Grand_Forks_1',  'knee_threshold': 0.85},
    {'company': 'IrisEnergy',     'knee_threshold': 0.85},
    {'company': 'Lancium',         'knee_threshold': 0.85},
    {'company': 'Marble_1',       'knee_threshold': 0.95},
    {'company': 'Marble_2',       'knee_threshold': 0.95},
    {'company': 'Mining Works',   'knee_threshold': 0.85},
    {'company': 'SATOtech',       'knee_threshold': 0.9},
    {'company': 'Soluna',         'knee_threshold': 0.85},
    {'company': 'StandardPower',  'knee_threshold': 0.9},
]

all_records = []

for idx, info in enumerate(knee_list, start=1):
    comp = info['company']
    knee_thr = info['knee_threshold']
    pattern = f"miner_{idx}_curtailment_results_*.csv"
    files = glob.glob(pattern)
    if not files:
        # Fallback to general pattern if specific threshold files are not found
        files = glob.glob(f"miner_{idx}_*.csv")
        if not files:
            continue
            
    # Use the first file found for this facility
    fn = files[0]

    # Try to load the data; skip if reading fails
    try:
        df = pd.read_csv(fn, parse_dates=['DateTime'])
    except Exception:
        continue
    
    # Try to extract threshold, default to 0.9
    try:
        thr_val = float(os.path.splitext(fn)[0].split('_')[-1]) / 100
    except (ValueError, IndexError):
        thr_val = 0.9 


    # --- Existing Metrics Calculation (from user's code) ---
    
    # basic counts
    total_hours      = len(df)
    dr_active_hours = df['DR_active'].sum()
    uptime_hours    = total_hours - dr_active_hours
    
    # 1. Uptime %  (non-DR hours)
    uptime_pct = uptime_hours / total_hours * 100 if total_hours>0 else np.nan
    
    # 2/3. Emissions sums
    avoided_em = df['avoided_emissions'].sum()
    induced_em = df['induced_emissions'].sum()

    # 4. Daily max energy
    max_energy = df['daily_max_energy'].max()

    # 5. DR relative % 
    dr_relative_pct = dr_active_hours / total_hours * 100 if total_hours>0 else np.nan

    # 6. LME variability
    lme_var = df['LME'].std() if 'LME' in df else np.nan

    if 'LME' in df and 'energy' in df and len(df) > 1:
        # 7. LME Skewness: Measures the asymmetry of LME distribution
        lme_skew = df['LME'].skew()
        
        # 8. LME-Curtailed Correlation: Pearson correlation between energy consumption and LME
        # A negative value means energy drops when LME is high (good alignment)
        lme_corr = df['energy'].corr(df['LME'])
        
        # 9. High LME Hour Density: Percentage of hours exceeding the facility's local 90th percentile LME
        lme_p90 = df['LME'].quantile(0.90)
        hlme_density = df.loc[df['LME'] > lme_p90].shape[0] / total_hours if total_hours > 0 else 0.0

    # 7. Energy savings & emission efficiency
    energy_savings = df['energy_savings'].sum() if 'energy_savings' in df else np.nan
    emission_eff   = (avoided_em / energy_savings
                         if energy_savings is not None and energy_savings > 0
                         else np.nan)

    # 8. Curtailment regularity
    if 'curtailment_start' in df and 'DateTime' in df:
        starts = df.loc[df['curtailment_start'], 'DateTime'].dt.time
        if len(starts)>1:
            secs = [t.hour*3600 + t.minute*60 + t.second for t in starts]
            curtail_reg = np.std(secs) / 3600.0
        else:
            curtail_reg = np.nan
    else:
        curtail_reg = np.nan

    # 9. Curtailment magnitude
    curtail_mag = calculate_curtailment_magnitude(df)

    # Normalized Induced emissions / hour of uptime
    norm_induced = induced_em / max_energy

    # 10. Facility ID (assumed constant in file)
    facility_id = df['Facility ID'].iloc[0] if 'Facility ID' in df and not df.empty else comp
    
    # --- START OF NEW METRICS ---

    # 11. Coefficient of Variation of Normal Load (CV-Normal)
    non_dr_energy = df.loc[~df['DR_active'], 'energy'].dropna()
    if len(non_dr_energy) > 1 and non_dr_energy.mean() != 0:
        cv_normal = non_dr_energy.std() / non_dr_energy.mean()
    else:
        cv_normal = 0.0

    # 12. Peak Load Alignment (PLA)
    pla = 0.0
    if 'LME' in df and dr_active_hours > 0 and len(df)>0:
        # 1. Find the 90th percentile LME (Top 10%) across *all* hours
        lme_threshold_90 = df['LME'].quantile(0.90)
        
        # 2. Count DR-active hours that are also above the threshold
        high_lme_dr_hours_count = df.loc[df['DR_active'] & (df['LME'] >= lme_threshold_90)].shape[0]
        
        # 3. Calculate PLA (ratio of strategic DR hours to total DR hours)
        pla = high_lme_dr_hours_count / dr_active_hours
        
    # 13. Average Curtailment Duration (ACD)
    acd = 0.0
    if dr_active_hours > 0 and 'DR_active' in df:
        # Identify groups of contiguous True values in DR_active
        # A new group starts when DR_active is True and the previous value was False
        start_of_new_event = df['DR_active'] & (~df['DR_active'].shift(1, fill_value=False))
        
        # Assign an ID to each event block
        event_id = start_of_new_event.cumsum()
        
        # Filter only for active DR hours and group by event_id
        dr_events = df.loc[df['DR_active']].copy()
        if not dr_events.empty:
            dr_events['event_id'] = event_id[df['DR_active']]
            
            # Calculate the duration of each event (counting hours in each block)
            event_durations = dr_events.groupby('event_id').size()
            
            # Calculate the mean duration
            acd = event_durations.mean()

    # peak load
    peak_load = df['energy'].max()

    # load factor
    avg_energy = df['energy'].mean()
    load_factor = avg_energy / peak_load
        
    # --- END OF NEW METRICS ---


    all_records.append({
        'Company': comp,
        'Facility ID': facility_id,
        'Threshold': thr_val,
        'Uptime (%)': uptime_pct,
        'DR Hours': dr_active_hours,
        'DR Relative (%)': dr_relative_pct,
        'Avoided Emissions (tCO)': avoided_em,
        'Induced Emissions (tCO)': induced_em,
        'Daily Max Energy (MWh)': max_energy,
        'LME Variability': lme_var,
        'LME Skewness': lme_skew,
        'LME-Curtailed Correlation': lme_corr,
        'HLME-Density (Local P90)': hlme_density,
        'Energy Savings (tCO)': energy_savings,
        'Emission Efficiency': emission_eff,
        'Curtailment Regularity (hr)': curtail_reg,
        'Curtailment Magnitude (%)': curtail_mag,
        'Normalized Induced Emissions (tCO/MWh)': norm_induced,
        'Knee Threshold': knee_thr,
        # Append NEW metrics
        'CV-Normal': cv_normal,
        'Peak Load Alignment (PLA)': pla,
        'Avg Curtailment Duration (hr)': acd,
        'Peak Load': peak_load,
        'Load Factor': load_factor
    })

# consolidate to a DataFrame
df_all = pd.DataFrame(all_records)

# 13. New derived metrics
df_all['Emissions Ratio'] = (
    df_all['Avoided Emissions (tCO)']
    / df_all['Induced Emissions (tCO)']
).replace([np.inf, -np.inf], np.nan)

df_all['Normalized Avoided Emissions (tCO/MWh)'] = (
    df_all['Avoided Emissions (tCO)']
    / df_all['Daily Max Energy (MWh)']
).replace([np.inf, -np.inf], np.nan)


# Display the last 5 rows and all columns, including the new ones
#print(df_all.tail().to_markdown(index=False))
df_all
# import os
# import glob
# import pandas as pd
# import numpy as np

# def calculate_curtailment_magnitude(df):
#     """
#     Calculates the % curtailment magnitude as:
#       100 * (1 - (avg_energy_in_lowest_quartile_of_DR_hours / baseline_avg_no_DR))
#     """
#     # 1) grab just the energy during DRactive hours
#     curtailment_hours = df.loc[df['DR_active'], 'energy'].dropna()

#     if curtailment_hours.empty:
#         # no DR hours => zero curtailment magnitude
#         return 0.0

#     # 2) find the 25th percentile of those DRhour energies
#     q25 = np.percentile(curtailment_hours, 25)
#     low_q = curtailment_hours[curtailment_hours <= q25]

#     # 3) average of that bottom quartile (or fall back to min if somehow empty)
#     avg_low_q = low_q.mean() if not low_q.empty else curtailment_hours.min()

#     # 4) baseline: mean energy during non-DR hours
#     baseline = df.loc[~df['DR_active'], 'energy'].mean()
#     if pd.isna(baseline) or baseline == 0:
#         # can't divide by zero or missing
#         return 0.0

#     # 5) percent curtailment magnitude
#     return 100.0 * (1.0 - (avg_low_q / baseline))

# # 1) List of facilities + knee thresholds
# knee_list = [
#     {'company': 'CPower',         'knee_threshold': 0.9},
#     {'company': 'Calvert_City_1', 'knee_threshold': 0.95},
#     {'company': 'Calvert_City_2', 'knee_threshold': 0.9},
#     {'company': 'Calvert_City_3', 'knee_threshold': 0.8},
#     {'company': 'CleanSpark',     'knee_threshold': 0.95},
#     {'company': 'Cormint',        'knee_threshold': 0.75},
#     {'company': 'Cottonwood_1',   'knee_threshold': 0.85},
#     {'company': 'Dalton_1',       'knee_threshold': 0.9},
#     {'company': 'Dalton_2',       'knee_threshold': 0.9},
#     {'company': 'Dalton_3',       'knee_threshold': 0.9},
#     {'company': 'Denton_DTO1',    'knee_threshold': 0.9},
#     {'company': 'Foreman',        'knee_threshold': 0.85},
#     {'company': 'Grand_Forks_1',  'knee_threshold': 0.85},
#     {'company': 'IrisEnergy',     'knee_threshold': 0.85},
#     {'company': 'Lancium',        'knee_threshold': 0.85},
#     {'company': 'Marble_1',       'knee_threshold': 0.95},
#     {'company': 'Marble_2',       'knee_threshold': 0.95},
#     {'company': 'Mining Works',   'knee_threshold': 0.85},
#     {'company': 'SATOtech',       'knee_threshold': 0.9},
#     {'company': 'Soluna',         'knee_threshold': 0.85},
#     {'company': 'StandardPower',  'knee_threshold': 0.9},
# ]

# all_records = []

# for idx, info in enumerate(knee_list, start=1):
#     comp = info['company']
#     knee_thr = info['knee_threshold']
#     pattern = f"miner_{idx}_curtailment_results_*.csv"
#     files = glob.glob(pattern)
#     if not files:
#         continue

#     for fn in files:
#         # extract threshold from filename
#         thr_val = int(os.path.splitext(fn)[0].split('_')[-1]) / 100

#         # load data
#         df = pd.read_csv(fn, parse_dates=['DateTime'])
        
#         # basic counts
#         total_hours     = len(df)
#         dr_active_hours = df['DR_active'].sum()
#         uptime_hours    = total_hours - dr_active_hours
        
#         # 1. Uptime %  (non-DR hours)
#         uptime_pct = uptime_hours / total_hours * 100 if total_hours>0 else np.nan
        
#         # 2/3. Emissions sums
#         avoided_em = df['avoided_emissions'].sum()
#         induced_em = df['induced_emissions'].sum()

#         # 4. Daily max energy
#         max_energy = df['daily_max_energy'].max()

#         # 5. DR relative %  
#         dr_relative_pct = dr_active_hours / total_hours * 100 if total_hours>0 else np.nan

#         # 6. LME variability
#         lme_var = df['LME'].std() if 'LME' in df else np.nan

#         # 7. Energy savings & emission efficiency
#         energy_savings = df['energy_savings'].sum() if 'energy_savings' in df else np.nan
#         emission_eff   = (avoided_em / energy_savings
#                           if energy_savings and energy_savings>0
#                           else np.nan)

#         # 8. Curtailment regularity
#         if 'curtailment_start' in df and 'DateTime' in df:
#             starts = df.loc[df['curtailment_start'], 'DateTime'].dt.time
#             if len(starts)>1:
#                 secs = [t.hour*3600 + t.minute*60 + t.second for t in starts]
#                 curtail_reg = np.std(secs) / 3600.0
#             else:
#                 curtail_reg = np.nan
#         else:
#             curtail_reg = np.nan

#         # 9. Curtailment magnitude
#         curtail_mag = calculate_curtailment_magnitude(df)

#         # Normalized Induced emissions / hour of uptime
#         #norm_induced = induced_em / uptime_hours
#         norm_induced = induced_em / max_energy

#         # 10. Facility ID (assumed constant in file)
#         facility_id = df['Facility ID'].iloc[0] if 'Facility ID' in df else comp

#         # 11. (later) knee threshold DR hours + relative %
#         #   well pick them out from the perthreshold list below

#         all_records.append({
#             'Company': comp,
#             'Facility ID': facility_id,
#             'Threshold': thr_val,
#             'Uptime (%)': uptime_pct,
#             'DR Hours': dr_active_hours,
#             'DR Relative (%)': dr_relative_pct,
#             'Avoided Emissions (tCO)': avoided_em,
#             'Induced Emissions (tCO)': induced_em,
#             'Daily Max Energy (MWh)': max_energy,
#             'LME Variability': lme_var,
#             'Energy Savings (tCO)': energy_savings,
#             'Emission Efficiency': emission_eff,
#             'Curtailment Regularity (hr)': curtail_reg,
#             'Curtailment Magnitude (%)': curtail_mag,
#             'Normalized Induced Emissions (tCO/MWh)': norm_induced,
#             'Knee Threshold': knee_thr,
#         })

# # consolidate to a DataFrame
# df_all = pd.DataFrame(all_records)

# # 13. New derived metrics
# df_all['Emissions Ratio'] = (
#     df_all['Avoided Emissions (tCO)']
#     / df_all['Induced Emissions (tCO)']
# ).replace([np.inf, -np.inf], np.nan)

# df_all['Normalized Avoided Emissions (tCO/MWh)'] = (
#     df_all['Avoided Emissions (tCO)']
#     / df_all['Daily Max Energy (MWh)']
# ).replace([np.inf, -np.inf], np.nan)


# df_all

# # 12. Extract DR hours at knee threshold, per facility
# #     we assume each (Company, Knee Threshold) pair is unique:
# knee_summary = (
#     df_all
#     .query("Threshold == Knee Threshold")
#     .loc[:, ['Company', 'DR Hours', 'DR Relative (%)']]
#     .rename(columns={
#         'DR Hours': 'DR at Knee (hrs)',
#         'DR Relative (%)': 'DR at Knee (%)'
#     })
# )

# # merge back onto full table if you like, e.g.:
# df_final = df_all.merge(
#     knee_summary,
#     on='Company',
#     how='left'
# )

# # 13. New derived metrics
# df_final['Emissions Ratio'] = (
#     df_final['Avoided Emissions (tCO)']
#     / df_final['Induced Emissions (tCO)']
# ).replace([np.inf, -np.inf], np.nan)

# df_final['Normalized Avoided Emissions (tCO/MWh)'] = (
#     df_final['Avoided Emissions (tCO)']
#     / df_final['Daily Max Energy (MWh)']
# ).replace([np.inf, -np.inf], np.nan)

# # sort & save
# df_final.sort_values(['Company','Threshold'], inplace=True)
# df_final.to_csv('facility_curtailment_metrics.csv', index=False)

# # show
# print(df_final.head(10))
```

```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

###
# --- 1. Setup and Helper Functions ---
###

# Assume company_names, dir_path, start_dt, and end_dt are pre-defined
# Example definitions:
# company_names = ['company_A', 'company_B']
# dir_path = './data'
# start_dt = pd.to_datetime('2023-07-01')
# end_dt   = pd.to_datetime('2023-09-30 23:59:59')

# Define your analysis window
start_dt = pd.to_datetime('2023-07-01')
end_dt   = pd.to_datetime('2023-09-30 23:59:59')


# Define the thresholds to test in each iteration
threshold_values = [i / 100.0 for i in range(0, 101, 5)]

def calculate_summary_curve(timeseries_df):
    """
    Takes a raw timeseries DataFrame and calculates the mean DR_percent for all thresholds.
    This function encapsulates the entire curtailment analysis logic.
    
    Args:
        timeseries_df (pd.DataFrame): DataFrame with 'DateTime', 'energy', and 'LME' columns.
        
    Returns:
        pd.DataFrame: A DataFrame with 'threshold' and 'mean_DR_percent' columns.
    """
    summary_records = []
    
    # Process for each threshold value
    for threshold in threshold_values:
        # Work on a copy to avoid modifying the original resampled df
        df = timeseries_df.copy()

        # Step 3: Extract additional columns and calculate derived metrics
        df['daily_max_energy'] = df.groupby(df['DateTime'].dt.date)['energy'].transform('max')
        
        def replace_outliers(df_in):
            grouped = df_in.groupby(df_in['DateTime'].dt.date)
            for date, group in grouped:
                prev_day = date - datetime.timedelta(days=1)
                if prev_day in grouped.groups:
                    prev_max_energy = grouped.get_group(prev_day)['daily_max_energy'].iloc[0]
                else:
                    prev_max_energy = 0
                outlier_threshold = 1.5 * prev_max_energy
                outliers = group['energy'] > outlier_threshold
                if outliers.any():
                    for idx in group.index[outliers]:
                        non_outliers_series = group.loc[~outliers, 'energy']
                        nearest_idx = non_outliers_series[non_outliers_series.index < idx].last_valid_index()
                        if pd.isna(nearest_idx):
                            nearest_idx = non_outliers_series[non_outliers_series.index > idx].first_valid_index()
                        df_in.at[idx, 'energy'] = df_in.at[nearest_idx, 'energy'] if nearest_idx is not None else group['daily_max_energy'].iloc[0]
            return df_in
        df = replace_outliers(df)
        
        # Step 4: Calculate DR_underclock_threshold and DR_flag
        df['daily_max_energy'] = df.groupby(df['DateTime'].dt.date)['energy'].transform('max')
        df['DR_underclock_threshold'] = threshold * df['daily_max_energy']
        df['DR_flag'] = df['energy'] < df['DR_underclock_threshold']
        
        # --- START: CORRECTED LOGIC FOR STEP 5 ---
        
        # Define a function to safely calculate the mean for non-DR hours within a group.
        # This function receives the entire group DataFrame for one day.
        def safe_avg_no_dr(group):
            # Filter for rows in this specific group that are NOT DR events
            energy_no_dr = group.loc[~group['DR_flag'], 'energy']
            # If a day has 100% DR events, the result is empty. Return NaN.
            # Otherwise, return the mean.
            return energy_no_dr.mean() if not energy_no_dr.empty else np.nan

        # 1. Use groupby.apply to get a single average value for each date.
        # This creates a mapping from each date to its calculated average.
        daily_avg_map = df.groupby(df['DateTime'].dt.date).apply(safe_avg_no_dr)

        # 2. Map these daily averages back to each row in the main DataFrame.
        df['energy_daily_avg_no_dr'] = df['DateTime'].dt.date.map(daily_avg_map)

        # 3. Handle any days that were 100% DR (now NaN) by filling them with the
        # value from the previous or next valid day.
        df['energy_daily_avg_no_dr'] = df['energy_daily_avg_no_dr'].ffill().bfill()
        
        # --- END: CORRECTED LOGIC FOR STEP 5 ---
        
        df['energy_savings'] = df['energy_daily_avg_no_dr'] - df['energy']
        df['DR_percent'] = (df['energy_savings'] / df['energy_daily_avg_no_dr'] * 100).fillna(0)
        df['DR_percent'] = df['DR_percent'].clip(lower=0)
        
        # Append the mean result for this threshold
        summary_records.append({
            'threshold': threshold,
            'mean_DR_percent': df['DR_percent'].mean()
        })
        
    return pd.DataFrame(summary_records)

def kneedle_max_distance(x, y, visualize=False):
    """Finds the knee of a curve by maximizing the distance to a line from start to end."""
    x, y = np.asarray(x), np.asarray(y)
    
    if np.max(x) == np.min(x) or np.max(y) == np.min(y) or len(x) < 3:
        return x[0] if len(x) > 0 else 0

    # Normalize data to the [0, 1] range
    x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
    y_norm = (y - np.min(y)) / (np.max(y) - np.min(y))

    # Line from first to last point
    xn0, yn0 = x_norm[0], y_norm[0]
    xn1, yn1 = x_norm[-1], y_norm[-1]

    # Line equation coefficients: a*x + b*y + c = 0
    a, b, c = yn1 - yn0, -(xn1 - xn0), xn1 * yn0 - yn1 * xn0
    
    denominator = np.sqrt(a**2 + b**2)
    if denominator == 0:
        return x[0]

    # Calculate perpendicular distances
    distances = np.abs(a * x_norm + b * y_norm + c) / denominator
    idx = np.argmax(distances)
    knee_x = x[idx]

    # Visualization logic (optional)
    if visualize:
        plt.figure(figsize=(8, 5))
        plt.plot(x_norm, y_norm, 'o-', label="Normalized DR (%) Curve")
        plt.plot([xn0, xn1], [yn0, yn1], 'r--', label="Baseline")
        plt.scatter(x_norm[idx], y_norm[idx], color='green', s=100, zorder=5, label="Knee")
        plt.xlabel("Normalized Threshold")
        plt.ylabel("Normalized Mean DR (%)")
        plt.title("Kneedle (Perp. Distance) with X & Y Normalization")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.savefig('kneedle_plot_normalized.png', dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()

    return knee_x

###
# --- 2. Main Bootstrapping Logic ---
###

n_bootstraps = 1000
confidence_level = 0.95
bootstrap_final_results = []
plot_generated = False

# Loop through each company to perform the analysis
for company_name in company_names:
    print(f"\n--- Processing company: {company_name} ---")
    try:
        # Load and merge the original energy and LME data ONCE
        company_dir_path = os.path.join(dir_path, company_name)
        energy_file_path = os.path.join(company_dir_path, 'energy_data.csv')
        lme_file_path = os.path.join(company_dir_path, 'lme_data.csv')

        energy_df = pd.read_csv(energy_file_path)
        lme_df = pd.read_csv(lme_file_path)
        
        # Convert to datetime and filter by date range
        energy_df['DateTime'] = pd.to_datetime(energy_df['DateTime'])
        lme_df['DateTime'] = pd.to_datetime(lme_df['DateTime'])
        energy_df = energy_df[(energy_df['DateTime'] >= start_dt) & (energy_df['DateTime'] <= end_dt)]
        lme_df = lme_df[(lme_df['DateTime'] >= start_dt) & (lme_df['DateTime'] <= end_dt)]
        
        # Keep essential columns and rename for consistency
        energy_df = energy_df[['DateTime', 'MWh']].rename(columns={'MWh': 'energy'})
        lme_df = lme_df[['DateTime', 'LME']]
        
        # Merge into a single original timeseries DataFrame
        df_original = pd.merge(energy_df, lme_df, on='DateTime', how='left').dropna()

        if df_original.empty:
            print(f"Skipping '{company_name}' due to no data in the specified date range.")
            continue
            
    except FileNotFoundError:
        print(f"Skipping '{company_name}' as data files were not found.")
        continue

    # 1. Calculate the knee on the ORIGINAL, non-resampled data
    print("Calculating knee for original data...")
    summary_curve_orig = calculate_summary_curve(df_original)
    x_orig = summary_curve_orig['threshold']
    y_orig = summary_curve_orig['mean_DR_percent']
    
    # Generate plot only for the first company with valid data
    visualize_this_run = not plot_generated
    original_knee = kneedle_max_distance(x_orig, y_orig, visualize=visualize_this_run)
    if visualize_this_run:
        plot_generated = True

    # 2. Perform bootstrapping
    bootstrapped_knees = []
    print(f"Starting {n_bootstraps} bootstrap iterations...")
    for i in range(n_bootstraps):
        # RESAMPLE THE ORIGINAL HOURLY DATA WITH REPLACEMENT
        df_resampled = df_original.sample(n=len(df_original), replace=True).reset_index(drop=True)

        # Recalculate the entire summary curve for the resampled data
        summary_curve_resampled = calculate_summary_curve(df_resampled)
        
        x_resampled = summary_curve_resampled['threshold']
        y_resampled = summary_curve_resampled['mean_DR_percent']

        # Find the knee for this bootstrap sample
        if len(x_resampled.unique()) > 2:
            knee = kneedle_max_distance(x_resampled, y_resampled, visualize=False)
            bootstrapped_knees.append(knee)
        
        if (i + 1) % 200 == 0: # Progress update
            print(f"  ...completed {i + 1}/{n_bootstraps} iterations.")

    # 3. Calculate confidence interval
    if bootstrapped_knees:
        lower_bound = np.percentile(bootstrapped_knees, (1 - confidence_level) / 2 * 100)
        upper_bound = np.percentile(bootstrapped_knees, (1 + confidence_level) / 2 * 100)

        bootstrap_final_results.append({
            'company': company_name,
            'original_knee_threshold': original_knee,
            'lower_ci': lower_bound,
            'upper_ci': upper_bound
        })
        print(f"Finished bootstrapping for {company_name}.")
    else:
        print(f"Could not compute bootstrap results for '{company_name}'.")

###
# --- 3. Display Final Results ---
###

if bootstrap_final_results:
    results_df = pd.DataFrame(bootstrap_final_results)
    print("\n--- Bootstrap Results for Knee Thresholds (95% CI) ---")
    print(results_df)
else:
    print("\nCould not compute bootstrap results for any company. Please check input data.")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, adjusted_rand_score, silhouette_score
import matplotlib.pyplot as plt # Added for plotting the silhouette scores (optional but helpful)

# Assuming uptime_df is already loaded and contains the necessary columns.

# --- 1) Select Features and Scale the Data ---
features_for_clustering = [
    'Normalized Induced Emissions (tCO/MWh)',
    'Load Factor',
    'Avg Curtailment Duration (hr)'
]

X = uptime_df[features_for_clustering].copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- NEW SECTION: 2) Find Best K using Silhouette Score ---
print("--- Finding Best K (Clusters) using Silhouette Score ---")

# Define a range of k values to test (e.g., 2 to 10)
k_range = range(2, 11)
silhouette_scores = []

for k in k_range:
    kmeans_model = KMeans(n_clusters=k, random_state=42, n_init='auto')
    cluster_labels = kmeans_model.fit_predict(X_scaled)

    # The silhouette score is undefined for a single cluster (k=1)
    if k > 1:
        score = silhouette_score(X_scaled, cluster_labels)
        silhouette_scores.append(score)
        print(f"k = {k}: Silhouette Score = {score:.4f}")

# Find the k with the maximum silhouette score
best_k_index = np.argmax(silhouette_scores)
best_k = k_range[best_k_index]
print(f"\nOptimal number of clusters (k) is: {best_k} (Max Silhouette Score: {silhouette_scores[best_k_index]:.4f})")
print("-" * 60)

# (Optional Plotting)
# plt.figure(figsize=(8, 4))
# plt.plot(k_range, silhouette_scores, marker='o')
# plt.title('Silhouette Scores for various k')
# plt.xlabel('Number of Clusters (k)')
# plt.ylabel('Silhouette Score')
# plt.show()


# --- 3) Perform K-Means Clustering with Best K ---
# Use the best_k found above for the main clustering
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init='auto')
uptime_df['Cluster'] = kmeans.fit_predict(X_scaled)


# --- 4) Sweep Thresholds from 0.00 to 1.00 in 0.05 steps ---
thresholds = np.round(np.arange(0.0, 1.0 + 1e-9, 0.05), 2)

rows = []
print("\nClustering vs. Threshold Grouping  Sensitivity to Cutoff")
print("=" * 62)
for t in thresholds:
    # Following your original convention:
    # Threshold Group = 1 if Knee Threshold < cutoff, else 0
    thr_group = (uptime_df['Knee Threshold'] < t).astype(int)

    # NOTE: The confusion matrix will now be size best_k x 2, if best_k > 2.
    # We continue assuming 2 groups for the threshold for the ARS.
    # For CM printing, we must use unique labels:
    unique_thr_labels = np.unique(thr_group)
    unique_clust_labels = np.unique(uptime_df['Cluster'])

    # ARS still works to compare the two labelings directly.
    ars = adjusted_rand_score(thr_group, uptime_df['Cluster'])

    # The confusion matrix printing/logic needs to be flexible for k > 2,
    # but the original code assumes 2x2. Let's adjust cm creation for flexibility
    # but keep the ARS as the main metric for comparison.

    # Re-run K-Means for the sweep section if you MUST keep the 2x2 structure (not recommended):
    # If the goal is strictly to compare the threshold (2 groups) against K-Means
    # results when K-Means also uses 2 clusters, then best_k should be ignored here.
    # Since the request was to find best K, we will use it and rely on ARS.

    # Only print CM if best_k is 2 to maintain the original table format.
    if best_k == 2:
        cm = confusion_matrix(thr_group, uptime_df['Cluster'], labels=[0, 1])
        total = cm.sum()
        correct = np.trace(cm)
        pct_agree = (correct / total * 100.0) if total else np.nan

        print(f"\nCutoff = {t:.2f}")
        print("Confusion Matrix (rows: Threshold Group 0/1, cols: Cluster 0/1):")
        print("              Cluster 0 | Cluster 1")
        print(f"Threshold Group 0: {cm[0]}")
        print(f"Threshold Group 1: {cm[1]}")
        print(f"Adjusted Rand Score: {ars:.4f}")
        print(f"Percent Agreement:   {pct_agree:.1f}%")

        rows.append({
            'k': best_k,
            'cutoff': t,
            'ARS': ars,
            'percent_agreement': pct_agree,
            'cm_00': cm[0, 0],
            'cm_01': cm[0, 1],
            'cm_10': cm[1, 0],
            'cm_11': cm[1, 1],
            'n_group0': int((thr_group == 0).sum()),
            'n_group1': int((thr_group == 1).sum()),
        })
    else:
        # If best_k > 2, the original CM printing/logic breaks. We must omit it.
        print(f"\nCutoff = {t:.2f}")
        print(f"Adjusted Rand Score: {ars:.4f} (Best K is {best_k}, CM is not 2x2)")
        rows.append({
            'k': best_k,
            'cutoff': t,
            'ARS': ars,
            'percent_agreement': np.nan,
            'cm_00': np.nan, 'cm_01': np.nan, 'cm_10': np.nan, 'cm_11': np.nan,
            'n_group0': int((thr_group == 0).sum()),
            'n_group1': int((thr_group == 1).sum()),
        })


sweep_results = pd.DataFrame(rows)


# --- 5) Identify the best cutoff by ARS (and break ties by percent agreement) ---
print("\n" + "=" * 62)
print(f"Best K-Means Result: k = {best_k}")
best = sweep_results.sort_values(['ARS', 'percent_agreement'], ascending=False).iloc[0]
print("Best cutoff by ARS (ties broken by percent agreement):")
print(f"Cutoff = {best['cutoff']:.2f} | ARS = {best['ARS']:.4f}")
if best['k'] == 2:
    print(f"Agreement = {best['percent_agreement']:.1f}%")
    print("Confusion Matrix at best cutoff:")
    print(f"[[{int(best['cm_00'])} {int(best['cm_01'])}]")
    print(f" [{int(best['cm_10'])} {int(best['cm_11'])}]]")
else:
    print("CM not displayed as k > 2.")


# (Optional) Sort the full table by ARS descending for quick review
sweep_results_sorted = sweep_results.sort_values('ARS', ascending=False)

# Display the table (in notebooks this will render nicely)
sweep_results_sorted
```

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt

# -------------------------
# 0) Config
# -------------------------
features_for_clustering = [
    'Normalized Induced Emissions (tCO/MWh)',
    'Emissions Ratio',
    'Uptime (%)',
]
k_values = list(range(2, 11))   # evaluate k=2..10
random_state = 42
n_init = 10                     # use 10 for broad sklearn compatibility
make_plots = True               # set False if you don't want plots

# -------------------------
# 1) Data & scaling
# -------------------------
X = uptime_df[features_for_clustering].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------
# 2) Fit across k and collect metrics
# -------------------------
records = []
for k in k_values:
    km = KMeans(n_clusters=k, random_state=random_state, n_init=n_init)
    labels = km.fit_predict(X_scaled)

    inertia = km.inertia_
    sil = silhouette_score(X_scaled, labels)            # higher is better
    ch = calinski_harabasz_score(X_scaled, labels)      # higher is better
    db = davies_bouldin_score(X_scaled, labels)         # lower is better

    records.append({
        "k": k,
        "inertia": float(inertia),
        "silhouette": float(sil),
        "calinski_harabasz": float(ch),
        "davies_bouldin": float(db),
    })

metrics_df = pd.DataFrame(records)

# -------------------------
# 3) Heuristic elbow detection via 2nd finite difference
#    (pick k with largest curvature drop)
# -------------------------
# inertia is defined for each k in ascending order
inertias = metrics_df["inertia"].values
# 2nd finite difference on inertia w.r.t. k (length = len(k_values)-2)
second_diff = np.diff(inertias, n=2)
# Elbow is where curvature (negative second diff) is largest -> maximize -second_diff
if len(second_diff) > 0:
    elbow_idx = int(np.argmax(-second_diff))  # index into second_diff array
    elbow_k = k_values[elbow_idx + 2]         # +2 because of second diff shrinkage
else:
    elbow_k = k_values[0]

# -------------------------
# 4) Best k per individual metric
# -------------------------
best_k_sil = int(metrics_df.loc[metrics_df["silhouette"].idxmax(), "k"])
best_k_ch  = int(metrics_df.loc[metrics_df["calinski_harabasz"].idxmax(), "k"])
best_k_db  = int(metrics_df.loc[metrics_df["davies_bouldin"].idxmin(), "k"])

# -------------------------
# 5) Simple majority vote across metrics (elbow, silhouette, CH, DB)
#    Tie-breakers: higher silhouette -> higher CH -> lower inertia
# -------------------------
votes = {k: 0 for k in k_values}
votes[elbow_k] += 1
votes[best_k_sil] += 1
votes[best_k_ch] += 1
votes[best_k_db] += 1

# find k with max votes
max_votes = max(votes.values())
candidates = [k for k, v in votes.items() if v == max_votes]

# tie-break using silhouette, then CH, then inertia (prefer smaller)
def tie_key(k):
    row = metrics_df.loc[metrics_df["k"] == k].iloc[0]
    return (
        -row["silhouette"],             # higher is better
        -row["calinski_harabasz"],      # higher is better
        row["inertia"]                  # lower is better
    )

best_k_overall = sorted(candidates, key=tie_key)[0]

# -------------------------
# 6) Output summary
# -------------------------
print("=== K-Means Model Selection (Usual Tests) ===")
print(f"Elbow (heuristic) suggests k = {elbow_k}")
print(f"Silhouette best k = {best_k_sil}")
print(f"CalinskiHarabasz best k = {best_k_ch}")
print(f"DaviesBouldin best k = {best_k_db}")
print(f"\nMajority-vote suggested k = {best_k_overall}")

print("\nMetrics table (higher Sil/CH better, lower DB/Inertia better):")
display(metrics_df)

# -------------------------
# 7) (Optional) Plots
# -------------------------
if make_plots:
    # Elbow (inertia)
    plt.figure()
    plt.plot(metrics_df["k"], metrics_df["inertia"], marker="o")
    plt.xlabel("k")
    plt.ylabel("Inertia (WCSS)")
    plt.title("Elbow Method (Inertia vs k)")
    plt.axvline(elbow_k, linestyle="--")
    plt.show()

    # Silhouette
    plt.figure()
    plt.plot(metrics_df["k"], metrics_df["silhouette"], marker="o")
    plt.xlabel("k")
    plt.ylabel("Average Silhouette Score")
    plt.title("Silhouette vs k")
    plt.axvline(best_k_sil, linestyle="--")
    plt.show()

    # CalinskiHarabasz
    plt.figure()
    plt.plot(metrics_df["k"], metrics_df["calinski_harabasz"], marker="o")
    plt.xlabel("k")
    plt.ylabel("CalinskiHarabasz")
    plt.title("CalinskiHarabasz vs k")
    plt.axvline(best_k_ch, linestyle="--")
    plt.show()

    # DaviesBouldin
    plt.figure()
    plt.plot(metrics_df["k"], metrics_df["davies_bouldin"], marker="o")
    plt.xlabel("k")
    plt.ylabel("DaviesBouldin (lower is better)")
    plt.title("DaviesBouldin vs k")
    plt.axvline(best_k_db, linestyle="--")
    plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm

df = uptime_df.copy()
df['group'] = df['Threshold Group']          # 0/1
df['size']  = df['Daily Max Energy (MWh)']

# --- Helper: partial Spearman via residualization (no extra packages) ---
def partial_spearman(x, y, controls):
    """
    Partial Spearman correlation between x and y controlling for 'controls' (array or DataFrame).
    Steps:
      1) Rank-transform x, y, and controls.
      2) Regress rank(x) on rank(controls); take residuals rx.
      3) Regress rank(y) on rank(controls); take residuals ry.
      4) Pearson corr between rx and ry == partial Spearman.
    Returns: r, p-value (two-sided).
    """
    # Ensure arrays
    x = np.asarray(x)
    y = np.asarray(y)
    Z = np.asarray(controls) if isinstance(controls, (np.ndarray, pd.Series)) else np.asarray(controls.values)

    # Rank-transform
    rx = stats.rankdata(x)
    ry = stats.rankdata(y)
    RZ = np.apply_along_axis(stats.rankdata, 0, Z) if Z.ndim == 2 else stats.rankdata(Z).reshape(-1, 1)

    # Add intercept
    RZ = sm.add_constant(RZ, has_constant='add')

    # OLS residuals (rank(x) ~ rank(controls))
    bx = sm.OLS(rx, RZ).fit()
    by = sm.OLS(ry, RZ).fit()
    rx_res = rx - bx.fittedvalues
    ry_res = ry - by.fittedvalues

    # Pearson on residuals == partial Spearman
    r = np.corrcoef(rx_res, ry_res)[0, 1]
    # p-value via t-test on correlation
    n = len(rx_res)
    t = r * np.sqrt((n - RZ.shape[1]) / (1 - r**2))  # df = n - k (k = #controls + intercept)
    from scipy.stats import t as tdist
    p = 2 * (1 - tdist.cdf(np.abs(t), df=n - RZ.shape[1]))
    return r, p

# Example: partial Spearman between avoided emissions and group, controlling for size
metrics_to_test = [
    'Avoided Emissions (tCO)',
    'Induced Emissions (tCO)',
    # add more raw or normalized metrics if you like
]

for metric in metrics_to_test:
    r, p = partial_spearman(df[metric], df['group'], df[['size']])
    print(f"Partial Spearman({metric} ~ group | size): r = {r:.3f}, p = {p:.4f}")

# OPTIONAL (if you have pingouin installed):
# import pingouin as pg
# for metric in metrics_to_test:
#     res = pg.partial_corr(data=df, x=metric, y='group', covar='size', method='spearman')
#     print(res)
```
